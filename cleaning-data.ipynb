```python
df = df.interpolate()
tmp = []
col = ['cctv1_jl-magelang-keluar-masuk', 'cctv1_jl-magelang-masuk']
tmp = df[col[0]].to_numpy() - df[col[1]].to_numpy()
df['cctv1_jl-magelang-keluar'] = tmp

data_a = df['cctv1_jl-magelang-masuk'].to_numpy()
data_b = df['cctv2_jl-magelang-masuk'].to_numpy()
data_c = df['cctv5_jl-magelang-masuk'].to_numpy()
tmp = []
for a, b, c in zip(data_a, data_b, data_c):
    _sum = 0
    div = 0
    if a is not None:
        _sum+=a
        div+=1
    if b is not None:
        _sum+=b
        div+=1
    if c is not None:
        _sum+=c
        div+=1
    _sum = int(_sum/div)
    tmp.append(_sum)
df['jl-magelang-masuk'] = tmp
df = df.drop(['cctv1_jl-magelang-keluar-masuk', 'cctv1_jl-magelang-masuk', 'cctv2_jl-magelang-masuk', 'cctv5_jl-magelang-masuk'], axis = 1)

data_a = df['cctv1_jl-magelang-keluar'].to_numpy()
data_b = df['cctv2_jl-magelang-keluar'].to_numpy()
data_c = df['cctv5_jl-magelang-keluar'].to_numpy()
tmp = []
for a, b, c in zip(data_a, data_b, data_c):
    _sum = 0
    div = 0
    if a is not None:
        _sum+=a
        div+=1
    if b is not None:
        _sum+=b
        div+=1
    if c is not None:
        _sum+=c
        div+=1
    _sum = int(_sum/div)
    tmp.append(_sum)
df['jl-magelang-keluar'] = tmp
df = df.drop(['cctv1_jl-magelang-keluar', 'cctv2_jl-magelang-keluar', 'cctv5_jl-magelang-keluar'], axis = 1)


col = ['cctv3_jl-kyaimojo-keluar', 'cctv3_jl-kyaimojo-keluar+masuk', ]
tmp = df[col[1]].to_numpy() - df[col[0]].to_numpy()
df['jl-kyaimojo-masuk'] = tmp
df = df.drop(['cctv3_jl-kyaimojo-keluar+masuk'], axis=1)



```

```python
data = df['cctv'].to_numpy()
tmp_date = []
tmp_hour = []
tmp_minutes = []
tmp_seconds = []
tmp_time = []


for idx in range(data.shape[0]):
    # _data = data[idx]
    datetime_str = data[idx][6:-4]
    try:
        datetime_object = datetime.strptime(datetime_str, '%Y-%m-%d %H:%M:%S.%f')
    except:
        print(datetime_str)
        break
    time_shift = timedelta(hours=7)
    datetime_object = datetime_object + time_shift
    date = datetime_object.date()
    hour = datetime_object.hour
    minutes = datetime_object.minute
    seconds = datetime_object.second
    tmp_date.append(f'{date}')
    tmp_hour.append(f'{hour}')
    tmp_minutes.append(f'{minutes}')
    tmp_seconds.append(f'{seconds}')
    tmp_time.append(f'{date}-{hour}:{minutes}')
df['date'] = tmp_date
df['hour'] = tmp_hour
df['minutes'] = tmp_minutes
df['seconds'] = tmp_seconds
df['time'] = tmp_time
```
